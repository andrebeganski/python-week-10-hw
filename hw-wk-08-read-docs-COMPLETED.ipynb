# Unstructured ----> Structured 

The zipped documents file for this week contains a mix of document types. For **only** the documents that hold the name of journalists and key data points associated with them, do the following:

1. Write a script that reads each relevant document and gathers:
- The Name of the journalist
- Article headline
- Publication
- Dateline

2. Export these data points to a ```.csv``` file


````python
## create more cells as needed.
import glob
import pandas as pd
````

````python
all_files = sorted(glob.glob("hw-wk-08-docs/some-docs-*.txt"))
journalist_dict = []
for file in all_files:
    with open(file, "r") as my_text:
        all_text_list = my_text.readlines()
        journalist_dict.append({
        "journalist name":all_text_list[0].replace("Name: ","").replace("\n",""),
        "article headline":all_text_list[2].replace("Article: ","").replace("\n",""),
        "publication":all_text_list[4].replace("Outlet: ","").replace("\n",""),
        "dateline":all_text_list[6].replace("Date: ","").replace("\n","")
        })
df = pd.DataFrame.from_dict(journalist_dict)
df.to_csv('journalist_data.csv', encoding = 'UTF-8', index = False)
df
````
